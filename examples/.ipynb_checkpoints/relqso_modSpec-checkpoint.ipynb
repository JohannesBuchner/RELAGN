# Example using relqso to create an SED

This is a quick introduction on how to use relqso to create a model SED
This is a simplified version of relagn, only taking a handful of parameters, and then calculating the rest (see the appendix in Hagen and Done (in prep.)) Also see the docs for a description of the input parameters and the class attributes/methods.

Throughout we will use units of photon/s/cm^2/keV, and keV (for examples on switching between unit outputs see the relagn_modSpec example)
We will start with the default input parameters, and then have a look at what changing mass accretion rate and spin does to the SED (as these will alter the derived coronal radius and spectral index)

We will also stick to using the relativistic version throuhghout. As in relagn this can be turned off (see the relagn_modSpec example on extracting non-relativistic spectra)

```python
#Importing packages
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

#adding relagn to pythonpath
import os
import sys
mdir = os.path.abspath('.')
mdir = mdir.replace('/examples', '/src/python_version')
sys.path.append(mdir)

from relagn import relqso
```

```python
rqso = relqso()
rqso.set_units('counts')
rqso.set_flux()

ph_tot = rqso.get_totSED()
ph_dsc = rqso.get_DiscComponent()
ph_wrm = rqso.get_WarmComponent()
ph_hot = rqso.get_HotComponent()

Es = rqso.E_obs
```

```python
#plotting
fig = plt.figure(figsize=(5, 6))
ax = fig.add_subplot(111)

ax.loglog(Es, Es**2 * ph_dsc, ls='-.', color='red')
ax.loglog(Es, Es**2 * ph_wrm, ls='-.', color='green')
ax.loglog(Es, Es**2 * ph_hot, ls='-.', color='blue')
ax.loglog(Es, Es**2 * ph_tot, color='k')

ax.set_ylim(max(Es**2 * ph_tot)*1e-2, max(Es**2*ph_tot)*1.5)
ax.set_xlim(1e-3, 1e3)

ax.set_xlabel('Energy   (keV)')
ax.set_ylabel(r'EF(E)   keV$^{2}$ (Photons s$^{-1}$ cm$^{-2}$ keV$^{-1}$)')
plt.show()

print('The calculated coronal parameters are:')
print(f'Gamma_h = {rqso.gamma_h}')
print(f'r_h = {rqso.r_h}')
print(f'r_w = {rqso.r_w}')
```

Above we have the SED for the defualt input parameters in relqso, along with the corresponding X-ray photon index, coronal radius, and outer radius of the warm region (as mentioned in Hagen & Done (in prep.) these are calculated by enforcing the condition that Lx_diss = 0.02Ledd in the frame of the black hole)

Now lets see what happens when we change the mass accretion rate
Remember that the model takes the log of the dimensionless mass accretion rate scaled by the Eddingon rate.
i.e $\dot{m} = \dot{M}/\dot{M}_{\mathrm{Edd}}$
Here we will run through the range $1.6 \leq \log \dot{m} \leq 0.4$

```python
lmdots = np.arange(-1.6, 0.5, 0.1) #np arange has final value as stop - step!
fig2 = plt.figure(figsize=(5, 6))
ax = fig2.add_subplot(111)

pmax = 0
cmap = mpl.cm.get_cmap('cool')
cnorm = mpl.colors.Normalize(-1.6, 0.4)
for i, lm in enumerate(lmdots):
    mqso = relqso(log_mdot=lm)
    mqso.set_units('counts')
    mqso.set_flux()
    
    ph_lm = mqso.get_totSED(rel=True)
    Es = mqso.E_obs
    
    ax.loglog(Es, Es**2 * ph_lm, color=cmap(cnorm(lm)), lw=1)
    
    imax = max(Es**2 * ph_lm)
    if imax > pmax:
        pmax = imax
    
ax.set_ylim(pmax*1e-4, pmax*1.5)
ax.set_xlim(5e-4, 1e3)

ax.set_xlabel('Energy   (keV)')
ax.set_ylabel(r'EF(E)   keV$^{2}$ (Photons s$^{-1}$ cm$^{-2}$ keV$^{-1}$)')

plt.colorbar(mpl.cm.ScalarMappable(norm=cnorm, cmap=cmap)).set_label(r'$\log \dot{m}$')
plt.show()
    
    
    
```

We can see that as we increase the mass accretion rate the X-ray photon index also increases (i.e it becomes softer). This is because as we increase $\dot{m}$ the power output from the accretion flow also increases. The requiremnt that $L_{x, \mathrm{diss}} = 0.02L_{\mathrm{Edd}}$ then forces the outer radius of the corona to reduce (because it now needs less area to reach the requered luminosity). Additionally the seed photon luminosity seen by the corona will increase, since there is more power output in the disc. An increase in seed photons leads to an increase in Compton cooling, which in turn leads to a softer photon index.

For more infor on the physics going on when setting $\Gamma_{h}$ see Kubota & Done (2018, https://ui.adsabs.harvard.edu/abs/2018MNRAS.480.1247K/abstract), which describes the original qsosed model (that relqso is based off). Also take a look at Mitchell et al (sub., https://ui.adsabs.harvard.edu/abs/2022arXiv221011977M/abstract), who give a robust test of qsosed (and also use relagn!)
