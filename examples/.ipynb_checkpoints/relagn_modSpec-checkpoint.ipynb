# Example on using relagn to create an SED
We will start by creating and plotting the SED corresponding to the model defualts (i.e defualt input parameters, and output units)

We will then play around with the output units and some parameters to see what happens

```python
#Importing packages
import numpy as np
import matplotlib.pyplot as plt

#adding relagn to pythonpath
import os
import sys
mdir = os.path.abspath('.')
mdir = mdir.replace('/examples', '/src/python_version')
sys.path.append(mdir)

from relagn import relagn
```

```python
dagn = relagn() #Default params, so no need to pass additional arguments. See the docs for defualt values

#Now extracting both relativistic and non-relativistic specrta
Lnu_rel = dagn.get_totSED(rel=True)
Lnu_nr = dagn.get_totSED(rel=False)

#Also extracting the spectral components (these are stored within the object when totSED is calculated)
#Hence no additional calculations needed!!!

#relativistic components
Lrel_dsc = dagn.get_DiscComponent()
Lrel_wrm = dagn.get_WarmComponent()
Lrel_hot = dagn.get_HotComponent()

#non-relativistic components
Lnr_dsc = dagn.get_DiscComponent(rel=False)
Lnr_wrm = dagn.get_WarmComponent(rel=False)
Lnr_hot = dagn.get_HotComponent(rel=False)
```

```python
#And now plotting. The output will be a single panel, containing both relativistic and non-relativistic SEDs
fig = plt.figure(figsize=(5, 6))
ax = fig.add_subplot(111)

nu = dagn.nu_obs #frequency grid
#plotting non-rel components first
ax.loglog(nu, nu*Lnr_dsc, ls='dotted', color='grey')
ax.loglog(nu, nu*Lnr_wrm, ls='dotted', color='grey')
ax.loglog(nu, nu*Lnr_hot, ls='dotted', color='grey')

ax.loglog(nu, nu*Lnu_nr, ls='dashed', color='grey')

#now doing rel components (in colour)
ax.loglog(nu, nu*Lrel_dsc, ls='-.', color='red')
ax.loglog(nu, nu*Lrel_wrm, ls='-.', color='green')
ax.loglog(nu, nu*Lrel_hot, ls='-.', color='blue')

ax.loglog(nu, nu*Lnu_rel, color='k')


ax.set_ylim(max(nu*Lnu_nr)*1e-2, max(nu*Lnu_nr)*2)
ax.set_xlim(5e14, 1e20)

ax.set_xlabel('Frequency   (Hz)')
ax.set_ylabel(r'$\nu F_{\nu}$   (ergs/s)')
plt.show()
```

Above shows the output SED (both with GR ray tracing (coloured lines/black solid line) and withour (grey lines) for the defualt relagn parameters. The red dahsed-dotted line shows the disc component, the green dashed-dotted line shows the warm Compton component, and the blue dashed-dotted line shows the hot Compton component.
The black solid line shoes the total SED in the full GR case, while the dashed grey line shows the full SED for the non-relativistic case.
Clearly there is very little difference between the non-relativistic and relativistic cases here. This is not entirely susrprising, as the default gives a 0 spin black hole; which in turn gives an innermost stable orbit, $r_{\mathrm{isco}} = 6$, and a moderate inclination of $\cos(i) = 0.5$
So the innermost part of the flow does not extend particularly close to the central black hole. Lets then try again, but this time for a maximally rotating blackl hole (i.e $a=0.998$, which gives $r_{\mathrm{isco}} \sim 1.24$

```python
sagn = relagn(a=0.998) 

#Now extracting both relativistic and non-relativistic specrta
Lnu_rel = sagn.get_totSED(rel=True)
Lnu_nr = sagn.get_totSED(rel=False)


#Also extracting the spectral components (these are stored within the object when totSED is calculated)
#Hence no additional calculations needed!!!

#relativistic components
Lrel_dsc = sagn.get_DiscComponent()
Lrel_wrm = sagn.get_WarmComponent()
Lrel_hot = sagn.get_HotComponent()

#non-relativistic components
Lnr_dsc = sagn.get_DiscComponent(rel=False)
Lnr_wrm = sagn.get_WarmComponent(rel=False)
Lnr_hot = sagn.get_HotComponent(rel=False)


#And now plotting. The output will be a single panel, containing both relativistic and non-relativistic SEDs
fig2 = plt.figure(figsize=(5, 6))
ax = fig2.add_subplot(111)

nu = sagn.nu_obs #frequency grid
#plotting non-rel components first
ax.loglog(nu, nu*Lnr_dsc, ls='dotted', color='grey')
ax.loglog(nu, nu*Lnr_wrm, ls='dotted', color='grey')
ax.loglog(nu, nu*Lnr_hot, ls='dotted', color='grey')

ax.loglog(nu, nu*Lnu_nr, ls='dashed', color='grey', label='non-rel')

#now doing rel components (in colour)
ax.loglog(nu, nu*Lrel_dsc, ls='-.', color='red')
ax.loglog(nu, nu*Lrel_wrm, ls='-.', color='green')
ax.loglog(nu, nu*Lrel_hot, ls='-.', color='blue')

ax.loglog(nu, nu*Lnu_rel, color='k', label='rel')


ax.set_ylim(max(nu*Lnu_nr)*1e-2, max(nu*Lnu_nr)*2)
ax.set_xlim(5e14, 1e20)

ax.set_xlabel('Frequency   (Hz)')
ax.set_ylabel(r'$\nu F_{\nu}$   (ergs/s)')

ax.legend(frameon=False)
plt.show()
```

Increasing spin both increases the power output at higher energies and the relativistic effects. The incease in X-ray power comes from an increase in the power dissipated in the hot corona due to $r_{\mathrm{isco}}$ reducing while $r_{h}$ has remained constant. The increase in relativistic effects is simply due to the closer proximitiy to the black hole. 

But what if I don't want the default output. Let's say, I would like the photon flux seen by the observer (e.g ph/s/cn^2/keV.
For this I can simply change the units and then re-etract the SED. Since the calculations have already been done internally you can do this without having to re-calculate the SED (it simply does the unit conversion during the extraction

```python
sagn.set_units('counts') #setting output to photon counts
sagn.set_flux() #Setting flux

#re-extracting
ph_rel = sagn.get_totSED()
ph_nr = sagn.get_totSED(rel=False)

Es = sagn.E_obs #energy grid

fig3 = plt.figure(figsize=(5, 6))
ax = fig3.add_subplot(111)

ax.loglog(Es, Es**2 * ph_nr, ls='dashed', color='grey', label='non-rel')
ax.loglog(Es, Es**2 * ph_rel, color='k', label='rel')

ax.set_ylim(max(Es**2 * ph_nr)*1e-2, max(Es**2 * ph_nr)*2)
ax.set_xlim(1e-3, 1e3)

ax.set_xlabel('Energy   (keV)')
ax.set_ylabel(r'EF(E)   keV$^{2}$ (Photons s$^{-1}$ cm$^{-2}$ keV$^{-1}$)')
ax.legend(frameon=False)
plt.show()

```

Or alternatively if you want the outpt in SI units per wavelength

```python
sagn.set_units('SI_wave')

Laa_rel = sagn.get_totSED()
Laa_nr = sagn.get_totSED(rel=False)

wave = sagn.wave_obs

fig4 = plt.figure(figsize=(5, 6))
ax = fig4.add_subplot(111)

ax.loglog(wave, wave*Laa_nr, ls='dashed', color='grey', label='non-rel')
ax.loglog(wave, wave*Laa_rel, color='k', label='rel')

ax.set_ylim(max(wave * Laa_nr)*1e-2, max(wave*Laa_nr)*2)
ax.set_xlim(1e-2, 1e4)

ax.set_xlabel('Wavelength   (Angstrom)')
ax.set_ylabel(r'$\lambda F_{\lambda}$   (W/$m^{2}$)')
ax.legend(frameon=False)
plt.show()
```
